```python
%matplotlib inline
import platform
print(platform.python_version())
import findspark
findspark.init()
import pyspark
from pyspark import SparkContext
print('\n' + ' //// Start my local Spark session //// ' + '\n')
```

```python
sc = SparkContext("local", "Simple Text App")
```

```python
sc.setLogLevel("WARN")
```

```python
sc.sparkUser()
```

```python
txtFile="../data/README.md"
txtData = sc.textFile(txtFile)
```

```python
lengthRDD = txtData.map(lambda x: len(x)).collect()
```

```python
print(lengthRDD)
```

Collect all pairs into new RDD

```python
# each line will be an x-value
pairsRDD = txtData.map(lambda x: (x, 1))
print(pairsRDD.collect() )
```

```python
# Split the sentences into words
pairsRDD = txtData.flatMap( lambda x: x.split(" ")).map(lambda x: (x, 1))
print(pairsRDD.collect())
```

reduce pairs by key

```python
pairsRDD.reduceByKey( lambda x,y: x+y).collect()
```

Excercise: reduce pairs by key and sort input in decending order

```python
pairsRDD.reduceByKey( lambda x,y: x+y).sortBy(lambda x: x[1], ascending=False).collect()
```

```python
#import numpy as np
import matplotlib.pyplot as plt

data=pairsRDD.reduceByKey( lambda x,y: x+y).sortBy(lambda x: x[1], ascending=False).take(20)
names = [x[0] for x in data]
x = range(len(data))
y = [x[1] for x in data]

print(names)
print(y)
# the histogram of the data

plt.bar(x, y, align='center', alpha=0.5)
plt.xticks(x, names)
plt.ylabel('frequencies')
plt.title('Programming language usage')
plt.show()
```
